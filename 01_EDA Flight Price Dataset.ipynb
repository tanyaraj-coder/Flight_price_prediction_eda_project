# EDA and Feature Engineering Flight Price Prediction 

## FEATURES 

The various features of the cleaned dataset are explained below:

1. Airline : The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines .
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

```python
#Plot count of flights by airline.

import matplotlib.pyplot as plt 

airline_counts = df['airline'].value_counts()

plt.figure()
plt.bar(airline_counts.index, airline_counts.values)
plt.xlabel('Airline')
plt.ylabel('Number of flights')
plt.title('Count of Flights by Airline')
plt.xticks(rotation=45)
for i, value in enumerate(airline_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom')
plt.show()

# airline_counts.values : This gives you: only the numbers (heights of the bars)
```

```python
# 3. plot average price per airline.

import pandas as pd 
import matplotlib.pyplot as plt

avg_price = df.groupby('airline')['price'].mean().sort_values()

# plot 

plt.figure(figsize=(10,6))
bars = plt.bar(avg_price.index, avg_price.values, color='pink', edgecolor='black')

# Add values on top of bars 

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 2), ha = 'center', va = 'bottom')

plt.title('Average price per Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')

plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt

count_dep = df['dep_period'].value_counts()
ax = count_dep.plot(kind='bar', color='lightgreen', edgecolor='black', figsize=(8,5))

for i, v in enumerate(count_dep):
    ax.text(i, v + 5, v, ha='center')

plt.title('Count of Flights by Departure Period')
plt.xlabel('Departure Period')
plt.ylabel('Number of Flights')
plt.show()
```

```python
# 5. Plot price vs deep_hour (scatter plot).

import matplotlib.pyplot as plt 

plt.figure(figsize=(10,6))

plt.scatter(df['dep_hour'], df['price'],color = 'orange', edgecolor='black', alpha= 0.7)
plt.title('Price vs Departure Hour')
plt.xlabel('Departure Hour')
plt.ylabel('Price')

plt.grid(True, linestyle='--', alpha = 0.5)

plt.show()
```

```python
# 6. Plot count of flights by number of stops .

import matplotlib.pyplot as plt

count_stops = df['Stops'].value_counts()

ax = count_stops.plot(kind= 'bar', color= 'violet', edgecolor= 'black',figsize=(8,5))

for i,v in enumerate (count_stops):
    ax.text(i,v+5,v,ha='center')

plt.title('Count of Flights by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Number of Flights')

plt.show()
```

+ i --> X position (which bar)
+ v --> Y valve (height of the bar = count)
+ v+5 --> puts the number a little above the bar ( so it doesn't overlap)
+ v --> the actual number to display
+ ha = 'center' --> keeps the number cetered on the bar .
